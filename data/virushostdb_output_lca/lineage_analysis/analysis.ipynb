```python
from pyecharts.charts import Pie
from pyecharts import options as opts


```

```python
import pandas as pd


        


def draw_pie_top20(df_, title,):
    df_ = df_[~df_.iloc[:, 0].isin(['-'])].copy()
    counts = list(zip(df_.iloc[:, 0], df_.iloc[:, 1]))
    word_counts = sorted(counts, key=lambda x: x[1], reverse=True)
    word_counts_top20 = top20(word_counts)

    # 绘图
    pie = (
        #创建一个饼图，初始化大小
        Pie(init_opts=opts.InitOpts(height="800px",width="1300px"))
        # 添加数据
        .add(

            "",
            word_counts_top20  # 数据
        )
        # 设置标签格式，b-名称，c是计数，d是概率
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}({d}%)",font_size=14))
        .set_global_opts(
            #标题
            title_opts=opts.TitleOpts(title=""),
            # 图例的布局朝向。可选：'horizontal', 'vertical'
            legend_opts=opts.LegendOpts(
                # 图例设置
                orient="vertical", pos_top="10%", pos_left='left', 
            ),
            #工具栏
            toolbox_opts=opts.ToolboxOpts(
                is_show=True,
                pos_top="bottom",
                pos_left="right",
                feature={
                    "saveAsImage": {},
                    "dataZoom": {"yAxisIndex": "none"},
                    "restore": {},
                    "magicType": {"show": True, "type": ["line", "bar"]},
                    "dataView": {}}
            )
        )

    )
    pie.render(f'{title}.html')  # 保存到本地
    pie.render_notebook()



def top20(word_counts):
    top20 = []
    if len(word_counts) > 20:
        for i in range(20):
            top20.append(word_counts[i])
        top20.append(('others', sum([x[1] for x in word_counts[20:]])))
        return top20
    else:
        return word_counts

# 1-kingdom
df=pd.read_table('1-kingdom.tsv', sep='\t')
draw_pie_top20(df,"1-kindom")
# 2-phylum
df=pd.read_table('2-phylum.tsv', sep='\t')
draw_pie_top20(df,"2-phylum")
# 3-class
df=pd.read_table('3-class.tsv', sep='\t')
draw_pie_top20(df,"3-class")
# 4-order
df=pd.read_table('4-order.tsv', sep='\t')
draw_pie_top20(df,"4-order")
# 5-family
df=pd.read_table('5-family.tsv', sep='\t')
draw_pie_top20(df,"5-family")
# 6-genus
df=pd.read_table('6-genus.tsv', sep='\t')
draw_pie_top20(df,"6-genus")
# 7-species
df=pd.read_table('7-species.tsv', sep='\t')
draw_pie_top20(df,"7-species")
# 8-strain
df=pd.read_table('8-strain.tsv', sep='\t')
draw_pie_top20(df,"8-strain")
```
